## **Rozkład prawdopodobieństwa**


**Rozkład Bernoulliego**

```python
import scipy.stats as scs
p=0.3
data=scs.bernoulli.rvs(p,size=10)
mean,var,skew,kurt=scs.bernoulli.stats(p,moments='mvsk')
print(mean)
print(var)
print(skew)
print(kurt)
```

```python
data
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams.update({'figure.figsize':(7,5), 'figure.dpi':100})

plt.hist(data, bins=2)
plt.gca().set(title='Częstość', ylabel='liczba wystąpień')
```

**Rozkład Poissona**

```python
import numpy as np
import matplotlib.pyplot as plt

mu=10
x=np.arange(scs.poisson.ppf(0.00001,mu),scs.poisson.ppf(0.99999,mu))
rv=scs.poisson(mu)
fig,ax=plt.subplots(1,1)
ax.vlines(x,0,rv.pmf(x),colors='b')
plt.show()
```

Wygeneruj dane z rozkładu normalnego. Policz średnią i odchylenie standardowe. Narysuj wykres rozkładu.

```python
from scipy.stats import norm
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)
mean, var, skew, kurt = norm.stats(moments='mvsk')
x = np.linspace(norm.ppf(0.01),
                norm.ppf(0.99), 100)
ax.plot(x, norm.pdf(x),
       'r-', lw=5, alpha=0.6, label='norm pdf')
rv = norm()
ax.plot(x, rv.pdf(x), 'k-', lw=2, label='frozen pdf')
```

```python
r = norm.rvs(size=1000)
```

```python
from scipy.stats import norm
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)
mu = 0
s=10
mean, var, skew, kurt = norm.stats(mu, s,moments= 'mvsk')
x = np.linspace(norm.ppf(0.01, mu,s), norm.ppf(0.99, mu,s), 100)
ax.plot(x, norm.pdf(x, mu,s), lw=5, alpha=0.6, label='norm pdf')
print(mean, var)
print(skew)
print(kurt)
```

Korzystając z testu Shapiro-Wilka sprawdź, czy podana próba ma 
rozkład normalny: [0.873, 2.817, 0.121, -0.945, -0.055, -1.436, 0.360, 
-1.478, -1.637, -1.869]. 

```python
normal_data =np.array([0.873, 2.817, 0.121, -0.945, -0.055, -1.436, 0.360, 
-1.478, -1.637, -1.869])
```

```python
ks_results = scs.kstest(normal_data, 'norm')

shapiro_results = scs.shapiro(normal_data)

print(ks_results)
print(shapiro_results)


plt.hist(normal_data, bins=50)
```
